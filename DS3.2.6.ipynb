```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
%matplotlib inline
```

```python
airbnb = pd.read_csv('/Users/salilketkar/thinkful/airbnb challenge/listings2.csv')
airbnbMod = airbnb.iloc[:,9:]
airbnbMod.head()
```

### Data Cleaning

```python
airbnbMod['price'] = airbnbMod['price'].apply(lambda x: str(x).replace('$',''))
airbnbMod['price'] = airbnbMod['price'].apply(lambda x: str(x).replace(',',''))
airbnbMod['price'] = airbnbMod['price'].apply(lambda x: str(x).replace('.',''))
airbnbMod['price'] = pd.to_numeric(airbnbMod['price'])


airbnbMod1 = airbnbMod.select_dtypes(include=['number'])
airbnbMod1.drop(['minimum_nights','maximum_nights','square_feet','jurisdiction_names','availability_30','availability_60','availability_90','availability_365'],axis=1,inplace=True)
airbnbMod1.dropna(inplace=True)


pd.set_option('display.max_columns',100)
airbnbMod1.head()
```

```python
airbnbData = airbnbMod1.drop('price',axis=1)
airbnbTarget = airbnbMod1.price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(airbnbData, airbnbTarget, test_size=0.3)
```

```python
from sklearn import tree

import timeit

start = timeit.default_timer()

decision_tree = tree.DecisionTreeRegressor(
max_features='auto',
min_samples_split=30
)

decision_tree.fit(X_train, y_train)
stop = timeit.default_timer()

print('Time: ', stop - start) 



print('Accuracy for train data:',decision_tree.score(X_train,y_train))
print('Accuracy for test data:',decision_tree.score(X_test,y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
import timeit

start = timeit.default_timer()

rf = RandomForestRegressor()
rf.fit(X_train,y_train)

stop = timeit.default_timer()

print('Time: ', stop - start) 


print('Accuracy for train data:',rf.score(X_train,y_train))
print('Accuracy for test data:',rf.score(X_test, y_test))
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(decision_tree, X_test, y_test, cv=10)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(rf, X_test, y_test, cv=10)
```
