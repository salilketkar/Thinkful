```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import data 
wine = pd.read_csv('/Users/salilketkar/thinkful/DS Unit 1/winequality-red.csv')
```

```python
wine.head()
```

```python
#choose one variable and plot four different ways

plt.hist(wine['pH'],bins=10)
plt.title('Histogram of Wine pH')
plt.xlabel('pH')
plt.show()

sns.set(style='whitegrid')
ax = sns.boxplot(x=wine['pH'])
plt.title('Boxplot of Wine pH')
plt.show()

norm = np.random.normal(0,1,1599)
norm.sort()
plt.plot(norm, wine['pH'].sort_values(), 'o')
plt.title('QQ plot comparing wine pH to normal distribution')
plt.ylabel('Wine pH')
plt.show()

plt.plot(wine['pH'])
plt.title('Line Graph for Wine pH')
plt.xlabel('Index')
plt.ylabel('Wine pH')
plt.show()
```

The histogram shows that the pH values are distributed normally.  The boxplot of wine pH shows potential outlier values, mean, and IQR.  The QQ plot compares the distribtuion of pH to a normal distribution of the same size.  The line graph doesn't provide any valuable information.  If there was data from a specific winery, about a specific type of wine, a line graph could be useful in showing how the pH changes over the years.  

```python
#choose two continious variables and plot three different ways (alcohol and pH)

sns.set(style='whitegrid')
g = sns.lmplot(y='pH', x='alcohol',data=wine,fit_reg=False,scatter_kws={'alpha':0.4})
g.set_ylabels('pH')
g.set_xlabels('Alcohol')
plt.title('Scatterplot: Correlation Between Alcohol and pH')
plt.show()

sns.set(style='ticks')
sns.jointplot(x='pH', y='alcohol',data=wine,kind="hex")
plt.show()

sns.set(style='whitegrid')
g = sns.lmplot(y='pH', x='alcohol',data=wine,fit_reg=True,scatter_kws={'alpha':0.4})
g.set_ylabels('pH')
g.set_xlabels('Alcohol')
plt.title('Scatterplot: Correlation Between Alcohol and pH')
plt.show()
```

The scatterplot shows the correlation between the pH and alcohol content in a wine.  The pH should increase as the alcohol content increases, making the wine more basic.  The second plot compares the same two variables using a hex jointplot.  The darker blue hexagons indicate that more data points fall at that particular pH and alcohol content.   The histograms along the top and right side show the distribution of the pH and alcohol content respectively.  The last plot is the same as the first but includes a regression line.  All three plots help in show trends between alcohol and pH.  The downside of these plots is that there are so many datapoints that these graphs are hard to interpret.   

```python
#Choose one continuous variable and one categorical variable, and plot them six different ways (quality=6,residual sugar)

wine5 = wine[wine['quality']==5]
wine6 = wine[wine['quality']==6]
wine7 = wine[wine['quality']==7]

plt.hist(wine5['residual sugar'],color='blue',alpha=0.5)
plt.hist(wine6['residual sugar'],color='green',alpha=0.5)
plt.hist(wine7['residual sugar'],color='red',alpha=0.5)
plt.title('Histogram of Residual Sugar with Wine Quality of 5, 6, and 7')
plt.xlabel('Residual Sugar')
plt.show()

newWine = pd.concat([wine5, wine6, wine7])
sns.set(style="darkgrid")
g = sns.factorplot(x='quality', y="residual sugar", data=newWine,size=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Residual Sugar")
g.set_xlabels("")
plt.title('Residual Sugar for Wines of Quality Score 5, 6, and 7')
plt.show()

sns.set(style="whitegrid")
g = sns.factorplot(x="quality", y="residual sugar", data=newWine,size=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Residual Sugar")
g.set_xlabels("")
plt.title('Pointplot: Residual Sugar for Wines of Quality Score 5, 6, and 7')
plt.show()
```

Three new dataframes were created with quality equal to 5, 6, and 7.  Quality is used as the categorical value for the three plots listed above.  The histogram shows the spread of residual sugar of wines with a quality score of 5, 6, and 7.  The barplot above shows the average residual sugar for wines with a quality score of 5, 6, and 7.  It also gives an error bar for each quality score.  A boxplot could be more appropriate in this situation since it provides more informaton about the variable.  The pointplot is a more visually pleasing way to see the changes residual sugar has across the categorical variable of quality.  

```python

sns.set(style="whitegrid")
ax = sns.boxplot(x='quality',y='residual sugar',data=newWine,palette='pastel')  
plt.title('Boxplots of Residual Sugar Levels for Wines of Quality Score 5, 6, and 7')
sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Residual Sugar')
plt.show()

sns.set(style="whitegrid")
g = sns.lmplot(y='residual sugar', 
               x='quality',
               data=newWine,
               fit_reg=False,
               ci=False,
               scatter_kws={'alpha':0.4})
g.set_ylabels("Residual Sugar")
g.set_xlabels("Quality")
plt.suptitle('Scatterplot Regression: Residual Sugar ')
plt.subplots_adjust(top=0.9)
plt.show()


```

The boxplot above shows the central tendency of residual sugar across the categorical quality variable of 5, 6, and 7.  It gives a good visualization of the outliers in all three boxplots.  The scatterplot regression is another way to visualize the spread of residual sugar across the quality variable.  It does not give any central tendency values though.
