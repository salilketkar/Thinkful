```python
#cleaning price column
journals.rename(columns={'COST () charged to Wellcome (inc VAT when charged)': 'Cost'}, inplace=True)
journals.dtypes
```

```python
journals['Currency'] = journals['Cost'].apply(lambda x:'EUR' if x.find("") >= 0 else 'USD')
journals['Cost'] = journals['Cost'].apply(lambda x: (str(x).replace('$','')))
journals['Cost'] = journals['Cost'].apply(lambda x: float(str(x).replace('','')))
journals.head()
```

```python
#Setting up conditions for np.select()
conditions = [
    (journals['Currency'] == 'USD'), 
    (journals['Currency'] == 'EUR'), 
    ]

#Picking target columns and apply operations to each number that satisfy the above 
choices = [journals['Cost']*0.86, journals['Cost']]

#Create a new column called "Converted_Salary" that contains all yearly salary. 
journals['convertedcost'] = np.select(conditions, choices, default=0)
```

```python
#removing prices listed as 0 euros and 999999 euros
updatedJournals = journals[(journals['convertedcost'] > 0) & (journals['convertedcost'] < 999999)]
```
