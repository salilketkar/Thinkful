```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#log distribution

dist1 = np.random.logseries(0.5, 100)
plt.hist(dist1)
plt.axvline(dist1.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(dist1.std() + dist1.mean(), color='c', linestyle='--', linewidth=2)
plt.axvline(dist1.mean() - dist1.std(), color='c', linestyle='--', linewidth=2)
plt.show()

print('The mean of dist1 is', np.mean(dist1))
print('The standard deviation of dist1 is', np.std(dist1))
```

```python
#negative binomial distribution
dist2 = np.random.negative_binomial(2, 0.5, 100)
plt.hist(dist2)
plt.axvline(dist2.mean(), color = 'b', linestyle= 'solid', linewidth = 2)
plt.axvline(dist2.mean() + dist2.std(), color = 'c', linestyle = '--', linewidth = 2)
plt.axvline(dist2.mean() - dist2.std(), color = 'c', linestyle = '--', linewidth = 2)
plt.show()

print('The mean of dist2 is', np.mean(dist2))
print('The standard deviation of dist2 is', np.std(dist2))
```

```python
#chisquared distribution
dist3 = np.random.chisquare(2, 100)
plt.hist(dist3)
plt.axvline(dist3.mean(), color = 'b', linestyle= 'solid', linewidth = 2)
plt.axvline(dist3.mean() + dist3.std(), color = 'c', linestyle = '--', linewidth = 2)
plt.axvline(dist3.mean() - dist3.std(), color = 'c', linestyle = '--', linewidth = 2)
plt.show()

print('The mean of dist3 is', np.mean(dist3))
print('The standard deviation of dist3 is', np.std(dist3))
```

```python
#standard normal distribution
dist4 = np.random.standard_normal(100)
plt.hist(dist4)
plt.axvline(dist4.mean(), color = 'b', linestyle= 'solid', linewidth = 2)
plt.axvline(dist4.mean() + dist4.std(), color = 'c', linestyle = '--', linewidth = 2)
plt.axvline(dist4.mean() - dist4.std(), color = 'c', linestyle = '--', linewidth = 2)
plt.show()

print('The mean of dist4 is', np.mean(dist4))
print('The standard deviation of dist4 is', np.std(dist4))
```

```python
#geometric distribution
dist5 = np.random.geometric(0.5, 100)
plt.hist(dist5)
plt.axvline(dist5.mean(), color = 'b', linestyle= 'solid', linewidth = 2)
plt.axvline(dist5.mean() + dist5.std(), color = 'c', linestyle = '--', linewidth = 2)
plt.axvline(dist5.mean() - dist5.std(), color = 'c', linestyle = '--', linewidth = 2)
plt.show()

print('The mean of dist5 is', np.mean(dist5))
print('The standard deviation of dist5 is', np.std(dist5))
```

```python
#pareto distribution
dist6 = np.random.pareto(10, 100)
plt.hist(dist6)
plt.axvline(dist6.mean(), color = 'b', linestyle= 'solid', linewidth = 2)
plt.axvline(dist6.mean() + dist6.std(), color = 'c', linestyle = '--', linewidth = 2)
plt.axvline(dist6.mean() - dist6.std(), color = 'c', linestyle = '--', linewidth = 2)
plt.show()

print('The mean of dist6 is', np.mean(dist6))
print('The standard deviation of dist6 is', np.std(dist6))
```

The standard normal distribution is the only distribution in which the mean and standard deviation are useful central measures.  The others present skewed data.

```python
rand1 = np.random.normal(5, 0.5, 100)
rand2 = np.random.normal(10, 1, 100)

rand3 = rand1 + rand2
plt.hist(rand3)
plt.axvline(rand3.mean(), color = 'b', linestyle= 'solid', linewidth = 2)
plt.axvline(rand3.mean() + rand3.std(), color = 'c', linestyle = '--', linewidth = 2)
plt.axvline(rand3.mean() - rand3.std(), color = 'c', linestyle = '--', linewidth = 2)
plt.show() 

print("Slightly skewed data.")
```
